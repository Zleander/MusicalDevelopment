# Data Preprocessing

In order to work with the Data effectively, a few preprocessing steps have to be taken.

Because the data  is very sparse in the years prior to 1990, in fact less than 5% of the datapoints lie in the timespan 1900-1989, from now on, we will only look at all datapoints with a release year of $\geq 1990$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import preprocessing
```

```python
# loading data
tracks = pd.read_csv('../data/tracks_features.csv', index_col=0)
# sorting out all datapoints with invalid year (=0) or year before 1990
tracks = tracks[tracks.year >= 1990].reset_index()
tracks
```

For further work with this dataset, we will split it into a train and a test dataset:

We will use the ratio of 80% train - and 20% test

```python
test_ratio = .2

no_total_points = len(tracks)
no_test_points = int(no_total_points * test_ratio)
no_train_points = no_total_points - no_test_points

print(f'We are using a ratio of {1-test_ratio} train, an {test_ratio} test \nthis results in {no_test_points} test and {no_train_points} training points')


train=tracks.sample(frac=0.8)
test=tracks.drop(train.index)
print(len(train), len(test))
assert len(train[train.id.isin(test.id)]) == 0, 'The train and test set overlap'
```

We now save these two Dataset for future use

```python
train.to_csv('../data/train_set.csv')
test.to_csv('../data/test_set.csv')
```
